1. Importing Open CV library

```python
import cv2 #importing open cv library
from google.colab.patches import cv2_imshow
```

2. Converting to grayscale using cvtcolor function

```python
pic=cv2.imread('/content/treee.jpg')#reading the frames of colored image
pic_gray=cv2.cvtColor(pic,cv2.COLOR_BGR2GRAY)#converting colored frames to grayscale
```

3. Shows the result

```python
cv2_imshow(pic) #shows colored img
cv2_imshow(pic_gray) #shows gray scale img
```
